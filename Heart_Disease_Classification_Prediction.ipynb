```python
high_corr_features = [
    "Pain_Arms_Jaw_Back", "Cold_Sweats_Nausea", "Dizziness", "Chest_Pain",
    "Fatigue", "Swelling", "Shortness_of_Breath", "Palpitations"
]
```

```python
dataframe_cleaned = dataframe.drop(high_corr_features, axis=1)
```

```python
X = dataframe_cleaned.drop(columns=['Heart_Risk'])
y = dataframe_cleaned['Heart_Risk']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
feature_names = X.columns  
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, log_loss
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
log_model = LogisticRegression(random_state=42,max_iter=1000)
log_model.fit(X_train_scaled, y_train)
log_importances = pd.Series(np.abs(log_model.coef_[0]), index=feature_names)

plt.figure(figsize=(10,5))
sns.barplot(
    x=log_importances.sort_values(ascending=False),
    y=log_importances.sort_values(ascending=False).index,
    palette='viridis'
)
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.title("Feature Importance ofLogistic Regression")
plt.show()
```

```python
gf_model = XGBClassifier(n_estimators=100, random_state=42)
gf_model.fit(X, y)

feature_importances = pd.Series(gf_model.feature_importances_, index=X.columns)
sorted_importances = feature_importances.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_importances, y=sorted_importances.index, palette="viridis")
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Feature Importance (Random Forest)")
plt.show()
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}
```

```python
results = {}
auc_scores = {}
log_losses = {}
predictions = {}
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:, 1]  
    predictions[name] = y_pred

    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)
    logloss = log_loss(y_test, y_proba)

    results[name] = accuracy
    auc_scores[name] = auc
    log_losses[name] = logloss

    print(f"Model: {name}")
    print(classification_report(y_test, y_pred))
    print("-" * 50)
    
    confusion_1 = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10,5))
    sns.heatmap(confusion_1, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
    plt.title('Confusion Matrix')
    plt.xlabel('predicted')
    plt.ylabel('real')
    plt.show()
```

```python
from sklearn.metrics import roc_auc_score
print('总的来看，xgb的模型更好')
param_dist = {  
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],
    'max_depth': [3, 5, 7, 9],
    'n_estimators': [100, 200, 300],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'scale_pos_weight': [1.0 , 1.5 ,1.9,2.1, 2.8]
}

xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=30,  
    cv=3,
    random_state=42,
    n_jobs=-1,  
    verbose=1
)

random_search.fit(X_train_scaled, y_train)
best_xgb = random_search.best_estimator_
y_pred_best = best_xgb.predict(X_test_scaled)
y_pred_proba_best = best_xgb.predict_proba(X_test_scaled)[:, 1]

print("Best Parameters:", random_search.best_params_)
print("\nXGBoost Tuned Report:\n", classification_report(y_test, y_pred_best))
print("\nXGBoost Tuned Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
y_pred_proba_best = best_xgb.predict_proba(X_test_scaled)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_proba_best)
print(f"\nXGBoost Tuned ROC AUC: {roc_auc}")
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
best_xgb = XGBClassifier(
    random_state=42,
    eval_metric=["logloss", "auc", "error"]  
)

eval_set = [(X_train_scaled, y_train), (X_test_scaled, y_test)]
best_xgb.fit(
    X_train_scaled, y_train,
    eval_set=eval_set,
    verbose=False
)

results = best_xgb.evals_result_

plt.figure(figsize=(6,4))
plt.plot(results['validation_0']['logloss'], label='Train LogLoss')
plt.plot(results['validation_1']['logloss'], label='Test LogLoss')
plt.legend(); plt.xlabel("Rounds"); plt.ylabel("LogLoss"); plt.title("XGBoost LogLoss"); plt.show()

plt.figure(figsize=(6,4))
plt.plot(results['validation_0']['auc'], label='Train AUC')
plt.plot(results['validation_1']['auc'], label='Test AUC')
plt.legend(); plt.xlabel("Rounds"); plt.ylabel("AUC"); plt.title("XGBoost AUC"); plt.show()

plt.figure(figsize=(6,4))
plt.plot([1-e for e in results['validation_0']['error']], label='Train Accuracy')
plt.plot([1-e for e in results['validation_1']['error']], label='Test Accuracy')
plt.legend(); plt.xlabel("Rounds"); plt.ylabel("Accuracy"); plt.title("XGBoost Accuracy"); plt.show()


precisions_train, recalls_train = [], []
precisions_test, recalls_test = [], []

for i in range(1, len(results['validation_1']['logloss'])+1):
    # 每一轮预测
    y_train_pred = (best_xgb.predict_proba(X_train_scaled, iteration_range=(0, i))[:,1] > 0.5).astype(int)
    y_test_pred  = (best_xgb.predict_proba(X_test_scaled, iteration_range=(0, i))[:,1] > 0.5).astype(int)
    
    # 存 precision / recall
    precisions_train.append(precision_score(y_train, y_train_pred))
    recalls_train.append(recall_score(y_train, y_train_pred))
    precisions_test.append(precision_score(y_test, y_test_pred))
    recalls_test.append(recall_score(y_test, y_test_pred))


plt.figure(figsize=(6,4))
plt.plot(precisions_train, label='Train Precision')
plt.plot(precisions_test, label='Test Precision')
plt.legend(); plt.xlabel("Rounds"); plt.ylabel("Precision"); plt.title("XGBoost Precision"); plt.show()

plt.figure(figsize=(6,4))
plt.plot(recalls_train, label='Train Recall')
plt.plot(recalls_test, label='Test Recall')
plt.legend(); plt.xlabel("Rounds"); plt.ylabel("Recall"); plt.title("XGBoost Recall"); plt.show()
```

```python
from sklearn.metrics import roc_curve
y_pred_proba = best_xgb.predict_proba(X_test_scaled)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = roc_auc_score(y_test, y_pred_proba)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, color='darkorange', lw=2,
         label=f'XGBoost ROC curve (area = {roc_auc:.3f})')
plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('ROC Curve for XGBoost')
plt.legend(loc="lower right")
plt.show()
```
